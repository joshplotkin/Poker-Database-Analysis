## General process

* ### Import first csv and make it the main csv
* ### Narrow down using the columns
* ### Import a new csv
* ### Join with main csv
* ### Narrow down again
* ### Repeat

```python
df = read_csv('4/16_batch01_4.csv', index_col = ('id_hand','id_player'))
del df['Unnamed: 0']

df = df[df.id_gametype == 1]
df = df[np.logical_or(np.logical_or(df.id_limit == 1,df.id_limit == 5),df.id_limit == 6)]
# limit 6 is 100NL
keep_index = df.index.copy()

df = df[['id_limit','position','id_holecard']]
```

```python
df.head(25)
```

```python
def table_join(df, stats, db, batch):
    df2 = read_csv(str(db) + '/16_batch' + batch + '_' + str(db) + '.csv', index_col = ('id_hand','id_player'))
    del df2['Unnamed: 0']
    df2 = df2[stats]
    df = df.join(df2)
    return df

df = table_join(df, ['cnt_players'], 4, '02')
df.head(25)
```
