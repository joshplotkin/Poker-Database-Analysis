Hampel's method:

find the Mean Absolute Deviation: 

$MAD = \frac{1}{N}\sum_{i=1}^N|x_i - \bar{X}|$

if a distance from median to data point is greater than 5.2 times the MAD, it's an outlier:

$|x - median| > 5.2MAD$

```python
MAD = abs(all_players['HR/FB_1'] - all_players['HR/FB_1'].median()).mean()
print all_players['HR/FB_1'].median() - MAD*5.2
print all_players['HR/FB_1'].median() - MAD*3
```

Tietjen-Moore test:

ratio of the sum of squares of non-outliers to outliers.

```python
k = 10

all_players.sort('HR/FB_1',inplace=True)
all_players['TEMP_NUMER'] = 0
## get mean of
for item in range(len(all_players.index)):
    if item >= k:
        all_players.iloc[item,8] = all_players.iloc[item,6]

for item in range(len(all_players.index)):
    if item < k:
        all_players.iloc[item,8] = all_players.TEMP_NUMER.mean()

tot_numer = ((all_players.TEMP_NUMER - all_players.TEMP_NUMER.mean())**2).sum()
tot_denom = ((all_players['HR/FB_1'] - all_players['HR/FB_1'].mean())**2).sum()
del all_players['TEMP_NUMER']

L = tot_numer / tot_denom
print L
```

Grubbs Test:

find the max Z-scores, and test again this statistic:

$$G = \frac{N-1}{N^{1/2}}\frac{t_{\alpha/N,N-2}}{\left(N-2+t_{\alpha/N,N-2}\right)^{1/2}}$$

```python
temp = all_players.copy()

N = len(min1000)
a = .05
t = stats.t.ppf(a/N,N-2)

G = (N-1)*t/(sqrt(N)*sqrt(N-2+t**2))
print G

## solve for threshold
print 'Grubbs Threshold: ' + str('%.3f ' % (G*min1000['HR/FB'].std() + min1000['HR/FB'].mean()))
```
