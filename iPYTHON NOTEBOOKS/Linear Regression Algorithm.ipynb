* ### Generate some points (with noise) along a regression line.

```python
lin_dict = {}

for i in range(26):
    lin_dict[i] = i + 2 + stats.norm.rvs(0,1,size=1)

df = DataFrame.from_dict(lin_dict, orient = 'index')
df['x'] = df.index
df['intercept'] = 1
df.columns = 'y','x','intercept'

fig = plt.figure(figsize = (12,8))
ax = fig.add_subplot(111)

colors = brewer2mpl.get_map('BrBG','diverging',4).mpl_colors

scatter(df.x,df.y, color = colors[3], s=50)

for label in ax.get_xticklines():
    label.set_visible(False)
for label in ax.get_yticklines():
    label.set_visible(False)

ax.set_xlabel('x')
ax.set_ylabel('y')
xlim(-.5,25.5)
ylim(0,30)

plt.show()
```

```python
fig = plt.figure(figsize = (12,8))
ax = fig.add_subplot(111)

colors = brewer2mpl.get_map('BrBG','diverging',4).mpl_colors

scatter(df.x,df.y, color = colors[3], s = 50)

x2 = np.arange(0,29,.1)
y_hat = res.params[0] + res.params[1]*x2

plot(x2,y_hat, color = colors[0], linewidth = 5)

for label in ax.get_xticklines():
    label.set_visible(False)
for label in ax.get_yticklines():
    label.set_visible(False)

ax.set_xlabel('x')
ax.set_ylabel('y')
xlim(-.5,25.5)
ylim(0,30)

plt.show()
```

* ### What if we introduce outliers?

```python
lin_dict = {}

for i in range(1,26):
    if i % 10 == 0:
        lin_dict[i] = stats.uniform.rvs(0,4,size=1)[0]
    else:
        lin_dict[i] = i + 2 + stats.norm.rvs(0,1,size=1)[0]


df = DataFrame.from_dict(lin_dict, orient = 'index')
df['x'] = df.index
df['intercept'] = 1
df.columns = 'y','x','intercept'
res = sm.OLS(df.y,df[['intercept','x']]).fit()
print res.summary()

fig = plt.figure(figsize = (12,8))
ax = fig.add_subplot(111)

colors = brewer2mpl.get_map('BrBG','diverging',4).mpl_colors

scatter(df.x,df.y, color = colors[3], s = 50)

x2 = np.arange(0,29,.1)
y_hat = res.params[0] + res.params[1]*x2

plot(x2,y_hat, color = colors[0], linewidth = 5)

for label in ax.get_xticklines():
    label.set_visible(False)
for label in ax.get_yticklines():
    label.set_visible(False)

ax.set_xlabel('x')
ax.set_ylabel('y')
xlim(-.5,25.5)
ylim(0,30)

plt.show()
```

* ### It doesn't look so bad but let's inspect the residuals.

```python
fig = plt.figure(figsize = (12,8))
ax = fig.add_subplot(111)

colors = brewer2mpl.get_map('BrBG','diverging',4).mpl_colors

scatter(df.x,df.y - res.predict(), color = colors[3], s = 50)

x2 = np.arange(0,29,.1)
y_hat = res.params[0] + res.params[1]*x2

plot(x2,np.zeros(len(x2)),color = colors[0], linewidth = 5)

for label in ax.get_xticklines():
    label.set_visible(False)
for label in ax.get_yticklines():
    label.set_visible(False)

ax.set_xlabel('x')
ax.set_ylabel('y')
# xlim(-.5,25.5)
# ylim(0,30)

ax.set_title('Residuals', fontsize = 20)

plt.show()
```

* ### These points have an affect on the $R^2$ value. What if we remove points that increase $R^2$ by 5% or more?
 * ### How can we handle this if we have many points?
 * ### How might we handle clusters of outliers?

```python
algo = df.copy()
while True:

    temp = algo.copy()
    res = sm.OLS(temp.y,temp[['intercept','x']]).fit()
    threshold = res.rsquared*1.05

    max_change = 0
    max_index = -1
    for i in temp.index:
        temp = algo.copy()
        temp = temp.drop(i)
        res = sm.OLS(temp.y,temp[['intercept','x']]).fit()
        if res.rsquared > threshold and res.rsquared > max_change:
            max_index = i
    if max_index == -1:
        break
    else:
        algo = algo.drop(max_index)

print algo
```

```python

res = sm.OLS(algo.y,algo[['intercept','x']]).fit()
print res.summary()

fig = plt.figure(figsize = (12,8))
ax = fig.add_subplot(111)

colors = brewer2mpl.get_map('BrBG','diverging',4).mpl_colors

scatter(algo.x,algo.y, color = colors[3], s = 50)

x2 = np.arange(0,29,.1)
y_hat = res.params[0] + res.params[1]*x2

plot(x2,y_hat, color = colors[0], linewidth = 5)

for label in ax.get_xticklines():
    label.set_visible(False)
for label in ax.get_yticklines():
    label.set_visible(False)

ax.set_xlabel('x')
ax.set_ylabel('y')
xlim(-.5,25.5)
ylim(0,30)

plt.show()
```
