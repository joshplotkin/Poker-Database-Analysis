* ### We create a time series of Poisson events. The parameter changes at day 40.

```python
count_data = read_csv("txtdata.csv", index_col = 'Day')


time_dict = {}

for i in range(75):
    if i < 40:
        time_dict[i] = stats.poisson.rvs(20,size=1)[0]
    else:
        time_dict[i] = stats.poisson.rvs(35,size=1)[0]

count_data = DataFrame.from_dict(time_dict, orient = 'index')

fig = plt.figure(figsize = (12,4))
ax = fig.add_subplot(111)

n_count_data = len(count_data)
plt.bar(np.arange(n_count_data), count_data.values, color="#348ABD")

ax.set_xlabel("Day #")
ax.set_ylabel("Occurrances")
ax.set_title("Time series data")
xlim(0, n_count_data);
```

* ### We use a Bayesian method to find the split point and parameters before and after.

```python
import pymc as pm

alpha = 1.0 / count_data.mean()  # Recall count_data is the
                               # variable that holds our txt counts
lambda_1 = pm.Exponential("lambda_1", alpha)
lambda_2 = pm.Exponential("lambda_2", alpha)

tau = pm.DiscreteUniform("tau", lower=0, upper=n_count_data)

print "Random output:", tau.random(), tau.random(), tau.random()

@pm.deterministic
def lambda_(tau=tau, lambda_1=lambda_1, lambda_2=lambda_2):
    out = np.zeros(n_count_data)
    out[:tau] = lambda_1  # lambda before tau is lambda1
    out[tau:] = lambda_2  # lambda after (and including) tau is lambda2
    return out

observation = pm.Poisson("obs", lambda_, value=count_data, observed=True)

model = pm.Model([observation, lambda_1, lambda_2, tau])

# Mysterious code to be explained in Chapter 3.
mcmc = pm.MCMC(model)
mcmc.sample(40000, 10000, 1)

lambda_1_samples = mcmc.trace('lambda_1')[:]
lambda_2_samples = mcmc.trace('lambda_2')[:]
tau_samples = mcmc.trace('tau')[:]

print ''
print 'Lambda for Time Period 1: ' + str('%.2f ' % lambda_1_samples.mean())
print 'Lambda for Time Period 2: ' + str('%.2f ' % lambda_2_samples.mean())
print 'Day the time changed: ' + str('%.2f ' % tau_samples.mean())
```

* ### What if we add outliers into the time series?

```python
count_data = read_csv("txtdata.csv", index_col = 'Day')
time_dict = {}

occurances = .25

for i in range(75):
    rand_num = stats.uniform.rvs(0,1,size=1)[0]
    if i < 40:
        if rand_num < occurances:
            print 'Outlier: ' + str(i)
            time_dict[i] = stats.uniform.rvs(30,35,size=1)[0]
        else:
            time_dict[i] = stats.poisson.rvs(20,size=1)[0]
    else:
        if rand_num < occurances:
            print 'Outlier: ' + str(i)
            time_dict[i] = stats.uniform.rvs(45,55,size=1)[0]
        else:
            time_dict[i] = stats.poisson.rvs(35,size=1)[0]

count_data = DataFrame.from_dict(time_dict, orient = 'index')

fig = plt.figure(figsize = (12,4))
ax = fig.add_subplot(111)

n_count_data = len(count_data)
plt.bar(np.arange(n_count_data), count_data.values, color="#348ABD")

ax.set_xlabel("Day #")
ax.set_ylabel("Occurrances")
ax.set_title("Time series data")
xlim(0, n_count_data);

```

* ### As expected, the parameters are thrown off by these outliers. Can we detect and remove the outliers, and find the underlying distribution?
 * ### Outliers are those that affect the sample mean by over 5%
 * ### Possible outliers affect the sample mean by over 1%

```python
count_data['OUTLIER'] = False
for i in count_data.index:
    if i < tau_samples.mean():
        if 1-stats.poisson.cdf(count_data.loc[i,0],lambda_1_samples.mean()) < 0.01:
            count_data.loc[i,'OUTLIER'] = True
        elif 1-stats.poisson.cdf(count_data.loc[i,0],lambda_1_samples.mean()) < 0.05:
            count_data.loc[i,'OUTLIER'] = 'Maybe'
    else:
        if 1-stats.poisson.cdf(count_data.loc[i,0],lambda_2_samples.mean()) < 0.01:
            count_data.loc[i,'OUTLIER'] = True
        elif 1-stats.poisson.cdf(count_data.loc[i,0],lambda_2_samples.mean()) < 0.05:
            count_data.loc[i,'OUTLIER'] = 'Maybe'


fig = plt.figure(figsize = (12,4))
ax = fig.add_subplot(111)
colors = brewer2mpl.get_map('Set1','qualitative',4).mpl_colors
            
inliers = count_data[count_data.OUTLIER == False]            
ax.bar(inliers.index.values, inliers[0].values, color=colors[1], label = 'Normal')

if len(count_data[count_data.OUTLIER == True] > 0):
    outliers = count_data[count_data.OUTLIER == True]
    ax.bar(outliers.index.values, outliers[0].values, color=colors[0], label = 'Outlier')

if len(count_data[count_data.OUTLIER == 'Maybe'] > 0):
    maybe = count_data[count_data.OUTLIER == 'Maybe']
    ax.bar(maybe.index.values, maybe[0].values, color=colors[3], label = 'Red Flag')



ax.set_xlabel("Day #")
ax.set_ylabel("Occurrances")
ax.set_title("Time series data")

# ax.legend(loc = 'best')
plt.xlim(0, n_count_data);
```

* ### We've discovered the outliers. Now we can remove them with the mean values of the non-outliers before and after the split point, respectively.

```python
inliers = count_data[count_data.OUTLIER != True]

x = int(round(tau_samples.mean()))
while True:
    if x in inliers.index:
        stop = x
        break
    else:
        x += 1

mean_under =  inliers.loc[:stop,0].mean()
mean_over = inliers.loc[stop:,0].mean()

print count_data

for i in count_data.index:
    if count_data.loc[i,'OUTLIER'] == True:
        if i < round(tau_samples.mean()):
            count_data.loc[i,0] = mean_under
        else:
            count_data.loc[i,0] = mean_over
del count_data['OUTLIER']
                
alpha = 1.0 / count_data.mean()  # Recall count_data is the
                               # variable that holds our txt counts
lambda_1 = pm.Exponential("lambda_1", alpha)
lambda_2 = pm.Exponential("lambda_2", alpha)

tau = pm.DiscreteUniform("tau", lower=0, upper=n_count_data)

@pm.deterministic
def lambda_(tau=tau, lambda_1=lambda_1, lambda_2=lambda_2):
    out = np.zeros(n_count_data)
    out[:tau] = lambda_1  # lambda before tau is lambda1
    out[tau:] = lambda_2  # lambda after (and including) tau is lambda2
    return out

observation = pm.Poisson("obs", lambda_, value=count_data, observed=True)

model = pm.Model([observation, lambda_1, lambda_2, tau])

mcmc = pm.MCMC(model)
mcmc.sample(40000, 10000, 1)

lambda_1_samples = mcmc.trace('lambda_1')[:]
lambda_2_samples = mcmc.trace('lambda_2')[:]
tau_samples = mcmc.trace('tau')[:]

print ''
print 'Lambda for Time Period 1: ' + str('%.2f ' % lambda_1_samples.mean())
print 'Lambda for Time Period 2: ' + str('%.2f ' % lambda_2_samples.mean())
print 'Day the time changed: ' + str('%.2f ' % tau_samples.mean())

fig = plt.figure(figsize = (12,4))
ax = fig.add_subplot(111)

n_count_data = len(count_data)
plt.bar(np.arange(n_count_data), count_data.values, color="#348ABD")

ax.set_xlabel("Day #")
ax.set_ylabel("Occurrances")
ax.set_title("Time series data")
xlim(0, n_count_data);
```
