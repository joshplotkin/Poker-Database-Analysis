```python
#hbk
df = hbk.copy()
df.Y = df.Y + df.Y.min() + .1

fig, ax = plt.subplots(3, 3, figsize = (14,14))

colors = brewer2mpl.get_map('Set2','qualitative',8).mpl_colors

fig.subplots_adjust(hspace=0.2)

color_list = (0,1,3)
list = df.columns
for i in range(3):
    for j in range(3):

        ax[i, j].spines['left'].set_visible(False)
        if i == j:
            points = ax[i, j].scatter(df[list[i]], df[list[j]], color = colors[color_list[i]]
                                      , alpha = .75, linewidths = 0, s = hbk.Y*50)
            ax[i, j].text((df[list[i]].max()-df[list[i]].min())/2,(df[list[i]].max()-df[list[i]].min())/2
                          ,list[i], fontsize = 20
                          , fontweight = 'bold', color = '#262626'
                          , alpha = 0.25)
            ax[i,j].set_xlabel(list[i])
        else:
            points = ax[i, j].scatter(df[list[i]], df[list[j]], color = colors[color_list[i]]
                                      , alpha = .75, linewidths = 0, s = hbk.Y*50)
            ax[i,j].set_xlabel(list[i])
            ax[i,j].set_ylabel(list[j])

plugins.connect(fig, plugins.LinkedBrush(points))
```

```python
#hbk
df = wood.copy()

fig, ax = plt.subplots(5, 5, figsize = (20,20))

colors = brewer2mpl.get_map('Set2','qualitative',8).mpl_colors

fig.subplots_adjust(hspace=0.5)
fig.subplots_adjust(wspace=0.5)

color_list = range(5)
list = df.columns
for i in range(5):
    for j in range(5):
        if i == j:
            points = ax[i, j].scatter(df[list[i]], df[list[j]], color = colors[color_list[i]]
                                      , alpha = .75, linewidths = 0, s = df.y*200)
            ax[i, j].text((df[list[i]].max()-df[list[i]].min())/2,(df[list[i]].max()-df[list[i]].min())/2
                          ,list[i], fontsize = 20
                          , fontweight = 'bold', color = '#262626'
                          , alpha = 0.25)
            ax[i,j].set_xlabel(list[i])
        else:
            points = ax[i, j].scatter(df[list[i]], df[list[j]], color = colors[color_list[i]]
                                      , alpha = .75, linewidths = 0, s = df.y*200)
            ax[i,j].set_xlabel(list[i])
            ax[i,j].set_ylabel(list[j])
        for label in ax[i,j].get_xticklabels():
            label.set_visible(False)
        for label in ax[i,j].get_yticklabels():
            label.set_visible(False)

plugins.connect(fig, plugins.LinkedBrush(points))
```
