situation:

villain bets nuts/air, pot-sized bet. 

we have hierarchical model of villain (ie. his bluff frequency), which we will show.

```python
pot = 100
bet = 100

bluff = .33333333

EV = (pot+bet)*bluff - bet*(1-bluff)
```

villain is distributed as Beta, with mean 35% bluffing.

beta parameters are normally distributed

all parameters are independent

```python
betarv = stats.beta.rvs(5,9,size=1000000)
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)

colors = brewer2mpl.get_map('Set1','qualitative',4).mpl_colors

hist(betarv, bins = 50, normed = True, histtype = 'bar', cumulative=False, color = colors[2], alpha = .5, label = 'Empirical Data')

plt.show()
```

```python
params = stats.beta.fit(betarv)
params
```
