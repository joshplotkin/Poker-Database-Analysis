```python
def make_stat_df(df, stats_list):

    df_list = ['HANDS','EV_250','EV_250_POST', 'LABEL', 'SET']
    
    for stat in stats_list:
        df_list.append(stat)
        df_list.append(stat+'_SMP')
    
    return df[df_list]
    
fold_to_cbet = make_stat_df(df, ['F2_F_CB_2B', 'F2_F_CB_3B','F2_T_CB_2B', 'F2_T_CB_3B','F2_R_CB_2B', 'F2_R_CB_3B'])
fold_to_cbet
```

```python
def plot_regression_bucket(df, hands, plot_types, degree, n):
    scatters = df.copy()

    sample = scatters.iloc[:,2]
    # might want to add min sample
    
    scatters = scatters.iloc[:,0:2]
    scatters.columns = 'XXXXX','YYYYY'

    print scatters.head()
    
    colors = brewer2mpl.get_map('YlGnBu', 'sequential', 8).mpl_colors

    if degree == 2:
        res = smf.ols('YYYYY ~ 1 + XXXXX + I(XXXXX**2.0)', scatters).fit()
    elif degree == 1:
        res = smf.ols('YYYYY ~ 1 + XXXXX', scatters).fit()
    else:
        print 'only handles 1 or 2 degrees.'
        
#     ax = plt.subplot(3,2,n)
    ax = plt.subplot(111)

#     xmin = 10
#     xmax = 65

    x = pd.DataFrame({'XXXXX': scatters.sort('XXXXX').XXXXX.values})
    xx = x.XXXXX.values

    if 'Data' in plot_types:
        scatter(scatters.XXXXX,scatters.YYYYY, color = colors[3],
                alpha = 0.5, label = 'Empirical Data')

    if 'Line' in plot_types:
        if 'Band' in plot_types:
            special.errorfill(xx, res.predict(x), res.bse[0]*2, label='Regression',
                      label_fill='Regression Error', color = '#0c2c84', alpha_fill = 0.1)
        else:
            plot(xx, res.predict(x), color = colors[7], linewidth = 3,
                 label = 'Regression Curve')



    xlabel('Standard Deviation', fontsize = 10)
    ylabel('BB/250', fontsize = 10)
    ax.set_title('Minimum ' + str(hands) + ' hands | $R^2$ = '
                 + str('% .3f ' % res.rsquared)
                 , fontsize = 10, fontweight = 'bold')


    if 'Fake Data' in plot_types:
        int_error =  res.bse[0]
        slope_error = res.bse[1]
        b0_ols = res.params[0]
        b1_ols = res.params[1]
        if degree == 2:
            b2_ols = res.params[2]
            poly_error = res.bse[2]

        xx = scatters.XXXXX.values
        yy = np.zeros(len(xx))
        for i in range(len(xx)):
            b0 = b0_ols + np.random.normal(0,int_error)
            b1 = b1_ols + np.random.normal(0,slope_error)
            yy[i] = b0+xx[i]*b1
            if degree == 2:
                b2 = b2_ols + np.random.normal(0,poly_error)
                yy[i] += xx[i]**2*b2

        plt.scatter(xx,yy, color = colors[7], label = 'Randomly Generated Data')

    for label in ax.get_xticklines():
        label.set_visible(False)
    for label in ax.get_yticklines():
        label.set_visible(False)
        
    xmin = scatters.XXXXX.min() 
    xmax = scatters.XXXXX.max()
    ymin = scatters.YYYYY.min()
    ymax = scatters.YYYYY.max()
    
    xlim([xmin,xmax])
    ylim([ymin,ymax])
#     ax.legend(loc = 'best')

fig = plt.figure(figsize = (20,16))
plt.suptitle('Is there a correlation between winrate and standard deviation?',
             fontsize = 16, color = '#262626', fontweight = 'bold')
plot_types = ('Data','Line','Band') #types:('Line','Data','Fake Data','Band')
degree = 1
n = 1
# for i in (350, 1000, 4000, 10000, 25000, 100000):
plot_regression_bucket(plot_df, i, plot_types, degree, n)
#     n += 1
```
