```python
import pandas as pd
import numpy as np
from pandas import *
import os
import psycopg2
import sys
import brewer2mpl
import pylab as pl
import matplotlib as mpl
import matplotlib
import matplotlib.pyplot as plt
import scipy as sc
from scipy import stats
from scipy.stats import norm
from pylab import *
from numpy import float64
from mpltools import style
from sklearn import mixture
from sklearn.cluster import KMeans

style.use('ggplot')

pd.set_option('max_rows',4000)
pd.set_option('max_columns',1000)
pandas.set_option('precision',3)
os.chdir('C:\Users\JP\Desktop\SCHOOL\Spring2014\EDAV\Poker data\NEW_TRY\StageE')

df = read_csv('14_ALL_STATS_WITH_SAMPLES.csv', index_col = 'NAME')
stat_names = {'WWSF': 'WWSF', 'WTSD': 'WTSD', 'WSD': 'W$@SD', 'WSD_exsm': 'W$@SD (excl. Small Pots)', 'WSD_rivcall': 'W$@SD (called river)', '3B': '3-bet', '4B': '4-bet', '4B_RNG': '4-bet range', 'FOLD_3B': 'Fold to 3-bet', 'FOLD_4B': 'Fold to 4-bet', 'CALL_4B': 'Call 4-bet', 'CALL_3B': 'Call 3-bet', 'COLDCALL': 'Coldcall', '3B_STEAL': '3-bet vs. steal', '3B_SUCCESS': '3-bet success', 'COLD_4B': 'Cold 4-bet', 'CB_F_2B': 'C-Bet Flop (Single Raised)', 'CB_F_3B': 'C-Bet Flop (3-bet Pot)', 'CB_T_2B': 'C-Bet Turn (Single Raised)', 'CB_T_3B': 'C-Bet Turn (3-bet Pot)', 'CB_R_2B': 'C-Bet River (Single Raised)', 'CB_R_3B': 'C-Bet River (3-bet Pot)', 'F2_F_CB_2B': 'Fold to Flop C-Bet (Single Raised)', 'F2_F_CB_3B': 'Fold to Flop C-Bet (3-bet pot)', 'F2_T_CB_2B': 'Fold to Turn C-Bet (Single Raised)', 'F2_T_CB_3B': 'Fold to Turn C-Bet (3-bet pot)', 'F2_R_CB_2B': 'Fold to River C-Bet (Single Raised)', 'F2_R_CB_3B': 'Fold to River C-Bet (3-bet pot)', 'R_F_CB_2B': 'Raise Flop C-Bet (Single Raised)', 'R_F_CB_3B': 'Raise Flop C-Bet (3-bet pot)', 'CR_F': 'Checkraise Flop', 'CR_T': 'Checkraise Turn', 'R_T_CB_2B': 'Raise Turn C-Bet (Single Raised)', 'R_T_CB_3B': 'Raise Turn C-Bet (3-bet pot)', 'F2R_CB_F_2B': 'Fold to Raise After Flop C-Bet (2B Pot)', 'F2R_CB_F_3B': 'Fold to Raise After Flop C-Bet (3B Pot))', 'F2R_CB_T_2B': 'Fold to Raise After Turn C-Bet (2B Pot)', 'F2R_CB_T_3B': 'Fold to Raise After Turn C-Bet (3B Pot))', 'AFq_TOT': 'Agg. Frequency Total', 'F_BVB': 'Fold to Raise BVB', '3B_BVB': '3-bet BVB', 'OPEN_SB': 'Open-raise SB', '3B_V_BTN': '3-bet vs. BTN', '3B_V_CO': '3-bet vs. CO', 'IFBET': 'Bet IP vs. Missed Flop C-Bet', 'ITBET': 'Bet IP vs. Missed Turn C-Bet', 'OTBET': 'Bet OOP vs. Missed Turn C-Bet', 'F2_IFBET': 'Flop Flop OOP After Not C-Betting', 'F2_ITBET': 'Flop Turn OOP After Not C-Betting', 'VPIP': 'VP$IP', 'PFR': 'PFR', 'PFR/VPIP': 'VP$IP:PFR', 'AFq_F': 'Agg. Frequency Flop', 'AFq_T': 'Agg. Frequency Turn', 'AFq_R': 'Agg. Frequency River', 'EV_250': 'EV/250', 'EV_250_POST': 'EV/250 (Posterior)', 'FOLD_3B_IP': 'Fold to 3-bet (IP)', 'FOLD_3B_OOP': 'Fold to 3-bet (OOP', 'FOLD_3B_STEAL': 'Fold to 3-bet (Steal)', 'RFI': 'Raise First In'}


def make_cluster_df(df,stat1,stat2,min_hands):
	cluster_df = df[[stat1,stat2,'HANDS','EV_250']]
	cluster_df = cluster_df[cluster_df.HANDS > min_hands]
	return cluster_df

def gmm_clustering(cluster_df,clusters):
	clf = mixture.GMM(n_components=clusters, covariance_type='full',random_state =123)
	clf.fit(cluster_df.iloc[:,0:2])
	train_predict = clf.predict(cluster_df.iloc[:,0:2])
	cluster_df['GMM'] = train_predict

	return cluster_df

def kmeans_clustering(cluster_df,clusters):
	est = KMeans(n_clusters=clusters, random_state =1241)
	est.fit(cluster_df.iloc[:,0:2])
	labels = est.labels_
	cluster_df['KMEANS'] = labels
	return cluster_df

def winrate_df(df,algo):
	df = df.groupby(algo).sum()
	df[algo] = df.BB_WON / df.HANDS * 250
	df[algo+'_HANDS'] = df.HANDS.apply(lambda x: "{:,.1f}M".format(float(x)/1000000))
	return df[[algo,algo+'_HANDS']]

def plot(cluster_df, winrates, proj_wr, stat1, stat2, num_clusters, min_hands, show):
	colors = brewer2mpl.get_map('YlGnBu','sequential',num_clusters).mpl_colors
	fig = plt.figure(figsize = (15,20))
	fig.suptitle(stat_names[stat1] + ' vs. ' + stat_names[stat2] + '\n' + str(len(cluster_df)) + ' players (regulars) | Min. ' + str(min_hands) + ' hands',
		fontsize = 14, color = '#262626', fontweight = 'bold')
	fig.subplots_adjust(hspace=0.4)

	xmin = cluster_df[stat1].min()
	ymin = cluster_df[stat2].min()
	xmax = cluster_df[stat1].max()
	ymax = cluster_df[stat2].max()

	cluster_dict = {1:'GMM',2:'KMEANS',3:'hexbin'}
	for n in cluster_dict:
		ax = fig.add_subplot(3,1,n)

		if n != 3:
			winrates.sort(cluster_dict[n], ascending = False, inplace = True)

			c = num_clusters-1
			for i in winrates.index:
				temp = cluster_df[cluster_df[cluster_dict[n]] == i]
				scatter(temp[stat1],temp[stat2],color = colors[c],
					label = str('%.2f' % (winrates.loc[i,cluster_dict[n]]/1.0)) + ' | ' + str(winrates.loc[i,cluster_dict[n]+'_HANDS'])) # divide by 2.5 for bb/100
				c -= 1

			ax.set_title(cluster_dict[n] + ' Clustering (Legend: Weighted BB/250 | Total Hands)',
				fontsize = 12)
			ax.set_xlabel(stat_names[stat1])
			ax.set_ylabel(stat_names[stat2])

			for label in ax.get_xticklines():
				label.set_visible(False)
			for label in ax.get_yticklines():
				label.set_visible(False)
			
			xlim(xmin,xmax)
			ylim(ymin,ymax)

			plt.legend(loc = 'best', bbox_to_anchor=(1.1, .99), fontsize = 10)

		if n == 3:
			x = proj_wr[stat1]
			y = proj_wr[stat2]
			z = proj_wr['PROJ_WR']
			hexplot = hexbin(x, y, C=z, gridsize=75, cmap=cm.PiYG, bins=None)
			cbar = plt.colorbar(hexplot, orientation='vertical', pad = 0, aspect = 10, label = 'Winrate')
			cbarytks = plt.getp(cbar.ax.axes, 'yticklines') # turn off tick lines
			plt.setp(cbarytks, visible=False)

			ax.set_title("Colors Represent Average Winrate of Players' Clusters",
				fontsize = 12)
			ax.set_xlabel(stat_names[stat1])
			ax.set_ylabel(stat_names[stat2])

			for label in ax.get_xticklines():
				label.set_visible(False)
			for label in ax.get_yticklines():
				label.set_visible(False)
			
			xlim(xmin,xmax)
			ylim(ymin,ymax)

	savefig('clusters/' + stat1 + '---' + stat2 + ' (clusters).png', bbox_inches='tight')
	if show == True:
		plt.show()

def plot_cluster(df,stat1,stat2,min_hands,num_clusters,show):
	df = make_cluster_df(df,stat1,stat2,min_hands)
	cluster_df = gmm_clustering(df,num_clusters)
	cluster_df = kmeans_clustering(cluster_df,num_clusters)

	cluster_df['BB_WON'] = cluster_df.HANDS * cluster_df.EV_250 / 250

	# get weight average winrates for the clusters
	winrates = winrate_df(cluster_df,'GMM').join(winrate_df(cluster_df,'KMEANS'))

	def gmm_assign_winrate(x):
		return winrates.loc[x,'GMM']
	def kmeans_assign_winrate(x):
		return winrates.loc[x,'KMEANS']

	cluster_df['GMM_WR'] = cluster_df.GMM.apply(gmm_assign_winrate)
	cluster_df['KMEANS_WR'] = cluster_df.KMEANS.apply(kmeans_assign_winrate)
	cluster_df['PROJ_WR'] = (cluster_df.GMM_WR+cluster_df.KMEANS_WR)/2
	proj_wr = cluster_df[[stat1,stat2,'PROJ_WR']]

	plot(cluster_df, winrates, proj_wr, stat1, stat2, num_clusters, min_hands, show)


# plot_cluster(df,'F2_F_CB_2B','F2_T_CB_2B',1,7)
plot_cluster(df,'PFR','CB_F_2B',250,7, False)

## TODO: add ability to zoom in by picking top 2 clusters


```
